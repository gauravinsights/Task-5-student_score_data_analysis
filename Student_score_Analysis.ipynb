### Drop/Delete Coloumn

```python
if "Unnamed: 0" in df.columns:
    df = df.drop("Unnamed: 0", axis=1)
```

```python
print(df.head())
```

### Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")  
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chat we have analysed that :
# The number of Female is more than the male.
```

### Relationship between Parents Education and Students score

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parents' Education and Students' Score")  
plt.show()
```

```python
# From the above chat we have concluded that the education of the parents have good impact on the students score.
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parent Marital Status   and Students' Score")  
plt.show()
```

```python
# from the above chat we have concluded that the martial status of the parents have No/Negligible impact on the students score.
```
